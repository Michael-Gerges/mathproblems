```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
chosen = []
a = [[0,1],[0,0],[1,1]]   # three cards brown-brown white-white and white-brown (white is 1 brown is zero)
B = [random.randint(0,2) for i in range(1000)] # create a list of the indecies that will be chosen 
thousand_cards = [a[B[i]] for i in range(1000)] # choose card based on the indecies list and append the 1000 card to a list 
```

```python
for i in range(1000):
    if sum(np.array(thousand_cards[i])) > 0: # out of the 1000 card remove a card with 0,0 which is equivelent to chose only the cards that one of there faces is white
        chosen.append(thousand_cards[i]) # create the new list chosen to have only the cards that one of their faces is white
```

```python
# if the possition of the face matters 
random_faces = np.array(chosen).flatten()
whites_ratio = np.array(random_faces).sum() / len(random_faces)
print(whites_ratio)
```

```python
count = 0
for i in range(len(chosen)):
    if  sum(np.array(chosen[i])) > 1: # look at the other face, is it white, how many was white, if both sides are white so the sum will be >2 
        count = count + 1
print(count/len(chosen)) # tell me how many cards you found with both white faces out of the cards that has posiblity that one of their faces is white  
```
