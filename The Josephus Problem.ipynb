```python

def letsdothis(name_list, num):
  alive = []
  for name in name_list:
    alive.insert(0,name)   # at the start of the list so it will be [5, 4, 3, 2, 1, 0]


  while len(alive) > 1:
    #print(alive) if you wanna see the survivors after each step
    for i in range(num+1):
       alive.insert(0,alive.pop() )# add the last dude (who just performed the execution) .. add him to the start of the list .. along with any skiped member if they wanna skip
    alive.pop()  # remove the executed member (if they dont skip (they execute the very next person): on the first round, it will be 2. Then, 4 then 6 then it comes back to number one who execute 3 then it goes to 5 who execute 1)
  print(alive[0])


letsdothis(list(range(1,8)), 0)  #the number person(s) they skip before they each execution
```
